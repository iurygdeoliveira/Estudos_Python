# Visualização de Dados - Evento Reestruture

```python
# Instalando bibliotecas para manipulação de dados
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import re
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from os import path
import matplotlib.pyplot as plt
```

## Carregando Dados

```python
dataset = pd.read_excel("Reestruture.xlsx")
```

```python
dataset
```

# Análise da Pergunta 01 - De 0 a 10 qual a sua avaliação sobre o evento

## Bubble Graph

```python
# Preparando os dados
```

```python
pclass = dataset['De 0 a 10 qual a sua avaliação sobre o evento'].value_counts().to_frame().reset_index().rename(columns = {'index':'De 0 a 10 qual a sua avaliação sobre o evento','De 0 a 10 qual a sua avaliação sobre o evento':'Count'})
```

```python
# Renderizando gráfico
```

```python
fig = go.Figure(data = [go.Scatter(x = pclass['De 0 a 10 qual a sua avaliação sobre o evento'], 
                                   y = pclass['Count'], 
                                   mode = 'markers', 
                                   marker = dict(size = pclass['Count']*3))])

# Layout
fig.update_layout(title = 'De 0 a 10 qual a sua avaliação sobre o evento ?',
                  xaxis_title = "Notas",
                  yaxis_title = "Número de Pessoas")

# Mostra o gráfico
fig.show()
```

## Pizza Graph

```python
# Preparando os dados
category = dataset['De 0 a 10 qual a sua avaliação sobre o evento'].value_counts().to_frame().reset_index().rename(columns = {'index':'De 0 a 10 qual a sua avaliação sobre o evento','De 0 a 10 qual a sua avaliação sobre o evento':'count'})
```

```python
#Renderizando Gráfico
```

```python
fig = go.Figure([go.Pie(labels = category['De 0 a 10 qual a sua avaliação sobre o evento'], values = category['count'])])

# Interatividade
fig.update_traces(hoverinfo = 'label+percent', 
                  textinfo = 'value+percent', 
                  textfont_size = 15,
                  insidetextorientation = 'radial')

# Layout
fig.update_layout(title = "De 0 a 10 qual a sua avaliação sobre o evento", title_x = 0.5)

# Gráfico
fig.show()
```

## Média

```python
mean_df = dataset['De 0 a 10 qual a sua avaliação sobre o evento'].mean()
print(mean_df)
```

# Análise da Pergunta 02 - O que mais você gostou ?

```python
#Preparando os dados
```

```python
dataset.columns
```

```python
words = dataset['O que mais você gostou ?']
words
```

```python
#Definindo a lista de stopwords
stopwords= set(STOPWORDS)

#Adicionando a lista stopwords em português
new_words = []
with open("stopwords_portuguese.txt", 'r') as f:
    [new_words.append(word) for line in f for word in line.split()]

new_stopwords = stopwords.union(new_words)
```

```python
%matplotlib inline
```

```python
#Renderizando Gráfico
```

```python
plt.figure(figsize=(20,10))
wc = WordCloud(min_font_size=10, 
               max_font_size=300, 
               background_color='white', 
               mode="RGB",
               stopwords=new_stopwords,
               width=2000, 
               height=1000,
               normalize_plurals= True).generate(' '.join(words))


plt.title("O que mais você gostou ?", fontsize=40, color="red")
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
